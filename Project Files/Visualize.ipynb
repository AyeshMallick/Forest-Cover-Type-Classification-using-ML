```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the original report data
data = {
    'Model': [
        'Logistic Regression',
        'K-Nearest Neighbors',
        'Decision Tree',
        'MLP Neural Networks'
    ],
    'Accuracy': [0.7268, 0.8972, 0.8188, 0.7278],
    'Precision': [0.7131, 0.8969, 0.8193, 0.7373],
    'Recall': [0.7268, 0.8972, 0.8164, 0.7278],
    'F1-Score': [0.7168, 0.8970, 0.8187, 0.7129]
}

# Load into a DataFrame and set the model names as the index
df = pd.DataFrame(data).set_index('Model')

# Transpose so that metrics are rows and models are columns
df_metrics = df.T

# Plot a grouped bar chart comparing models for each metric
ax = df_metrics.plot(kind='bar')
ax.set_title('Comparison of All Models Based on Performance Metrics')
ax.set_xlabel('Metric')
ax.set_ylabel('Score')
ax.set_ylim(0, 1)
plt.xticks(rotation=0)
plt.legend(title='Model', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```
